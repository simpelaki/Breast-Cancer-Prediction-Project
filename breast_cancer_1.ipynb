```python
#import libraries

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
from pprint import pprint

```

```python
#Draw the scatter plot matrix to explore the raltionship between variables

grid = sns.PairGrid(data.iloc[:, 0:11],palette="Set2",hue="diagnosis")
grid = grid.map_diag(plt.hist, edgecolor="b")
grid = grid.map_offdiag(plt.scatter, edgecolor="b", s=40)
plt.show()
```

```python
#Plot the histogram of area mean

area_mean_m = data[data['diagnosis'] ==1]['area_mean']
area_mean_b = data[data['diagnosis'] ==0]['area_mean']

plt = go.Figure()
plt.add_trace(go.Histogram(x=area_mean_b, name = 'Benign', marker_color='purple'))
plt.add_trace(go.Histogram(x=area_mean_m, name = 'Malignant', marker_color='orange'))

# set the mode as overlay
plt.update_layout(
    title = 'Area_mean Comparison of Benign & Malignant', 
                  xaxis_title ='Area_Mean Value ->',
                  yaxis_title = 'Value Counts ->',
                  barmode='overlay'
             
)

# Reduce opacity to see both histograms
plt.update_traces(opacity=0.75)
plt.show()

pprint(f'Mean of area values (Benign): {area_mean_b.mean()}')
pprint(f'Mean of area values (Malignant): {area_mean_m.mean()}')
```
